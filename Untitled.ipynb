```python
def staBar(br):

    (Path.cwd() /'result'/'does'/br/'graph'/'stat').mkdir(parents=True, exist_ok=True)
    statDoes = pd.read_pickle(Path.cwd() /'result'/'does'/br/'statDoes.p') #upload

    #variables
    width = 0.46 #float
    colors = plt.cm.BuPu(np.linspace(0.2, 0.5, 3))  #np.array
    xlines = np.linspace(9., 16.5, len(statDoes.columns.levels[0])) #xvalues #len 16
    print (xlines)
    month = statDoes.index

    for im in range(0, len(month)):

       #Prepare
        fig = plt.figure()
        meanlines = statDoes.iloc[im, ::2].values
        stdlines =  statDoes.iloc[im, 1::2].values
        
        xlabel = statDoes.columns.levels[0] #[0900-0930, 0930-1000,..., 1630-1700]
        xloc = np.linspace(9., 16.5, 16) #[9.0, 9.5, 10, ..., 16.5]
        
        fig.set_size_inches([3.23,2.59])
        plt.bar(xlines, meanlines,width=width,yerr=stdlines, align='center',color=colors[1], label='mean')
        
        plt.title('Mean Does +-1 stddev on ' + month[im])
        plt.xticks(xloc, xlabel, rotation='vertical')
        plt.xlabel('Time [Hour]'); plt.ylabel('Does[SAPD]')
        
        plt.axhline(1.0, label='1.0 SAPD',linestyle='--', color='r')
        plt.legend(bbox_to_anchor=(0.75, .7, 1., .102), loc=3, fontsize='small')

        fname = month[im] + '.jpg'
        if not (Path.cwd() /'result'/'does'/br/'graph'/'stat'/fname).exists():
            plt.savefig(Path.cwd() / 'result'/'does'/br /'graph'/'stat'/fname, bbox_inches='tight')
            print (fname)
            
        plt.show()
```
